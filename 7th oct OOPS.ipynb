```python
# PROBLEM -1

class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited {amount} into account {self.account_number}. New balance: {self.balance}"
        else:
            return "Invalid deposit amount."

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            return f"Withdraw {amount} from account {self.account_number}. New balance: {self.balance}"
        else:
            return "Invalid withdrawal amount or insufficient balance."

    def display_balance(self):
        return f"Account {self.account_number} balance: {self.balance}"


account1 = BankAccount("12345", "vicky")
print(account1.deposit(1000))
print(account1.withdraw(500))
print(account1.display_balance())
```

```python
#2 PROBLEM-2

class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        bonus = (bonus_percentage / 100) * self.salary
        return bonus

    def display_details(self):
        return f"Employee ID: {self.employee_id}, Name: {self.name}, Salary: {self.salary}"


employee1 = Employee("E001", "vicky", 50000)
print(f"Yearly Bonus: ${employee1.calculate_yearly_bonus(10)}")
print(employee1.display_details())
```

```python
# PROBLEM -3
class VehicleRental:
    def __init__(self):
        self.available_vehicles = {}

    def add_vehicle(self, vehicle_id, vehicle_name):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_name
            return f"Added {vehicle_name} (ID: {vehicle_id}) to available vehicles."
        else:
            return "Vehicle with that ID already exists."

    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            vehicle_name = self.available_vehicles.pop(vehicle_id)
            return f"Rented {vehicle_name} (ID: {vehicle_id})."
        else:
            return "Vehicle not available for rent."

    def return_vehicle(self, vehicle_id, vehicle_name):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_name
            return f"Returned {vehicle_name} (ID: {vehicle_id})."
        else:
            return "Invalid return: Vehicle already available."

    def display_available_vehicles(self):
        return "Available Vehicles:\n" + "\n".join([f"ID: {k}, Vehicle: {v}" for k, v in self.available_vehicles.items()])


rental_system = VehicleRental()

print(rental_system.add_vehicle(1, "Car"))
print(rental_system.add_vehicle(2, "Bike"))
print(rental_system.add_vehicle(3, "Scooter"))

print(rental_system.display_available_vehicles())

print(rental_system.rent_vehicle(2))
print(rental_system.display_available_vehicles())

print(rental_system.return_vehicle(2, "Bike"))
print(rental_system.display_available_vehicles())
```

```python
#4 PROBLEM - 4
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.available = True

    def __str__(self):
        return f"Book ID: {self.book_id}\nTitle: {self.title}\nAuthor: {self.author}\nAvailable: {'Yes' if self.available else 'No'}"

    def borrow(self):
        if self.available:
            self.available = False
            return f"You have borrowed '{self.title}' by {self.author}."
        else:
            return f"'{self.title}' is already borrowed."

    def return_book(self):
        if not self.available:
            self.available = True
            return f"You have returned '{self.title}' by {self.author}. Thank you!"
        else:
            return f"'{self.title}' is not borrowed, so you can't return it."


class LibraryCatalog:
    def __init__(self):
        self.books = {}

    def add_book(self, book_id, title, author):
        if book_id not in self.books:
            self.books[book_id] = Book(book_id, title, author)
            return f"Added '{title}' by {author} (ID: {book_id}) to the library catalog."
        else:
            return "Book with that ID already exists in the library catalog."

    def display_available_books(self):
        available_books = [book for book in self.books.values() if book.available]
        if available_books:
            return "Available Books:\n" + "\n\n".join([str(book) for book in available_books])
        else:
            return "No books are available in the library catalog."

library = LibraryCatalog()

print(library.add_book(1, "The Great Gatsby", "F. Scott Fitzgerald"))
print(library.add_book(2, "To Kill a Mockingbird", "Harper Lee"))
print(library.add_book(3, "1984", "George Orwell"))

print(library.display_available_books())

book = library.books[1]
print(book.borrow())

print(book.return_book())

print(library.display_available_books())
```

```python
#5 PROBLEM - 5

class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: ${self.price:.2f}\nQuantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product_id, name, price, quantity):
        if product_id not in self.products:
            self.products[product_id] = Product(product_id, name, price, quantity)
            return f"Added {quantity} units of '{name}' (ID: {product_id}) to the inventory."
        else:
            return "Product with that ID already exists in the inventory."

    def update_quantity(self, product_id, quantity_change):
        if product_id in self.products:
            self.products[product_id].quantity += quantity_change
            return f"Updated quantity of '{self.products[product_id].name}' (ID: {product_id}) to {self.products[product_id].quantity}."
        else:
            return "Product with that ID does not exist in the inventory."

    def display_available_products(self):
        available_products = [product for product in self.products.values() if product.quantity > 0]
        if available_products:
            return "Available Products:\n" + "\n\n".join([str(product) for product in available_products])
        else:
            return "No products are available in the inventory."

inventory = Inventory()

print(inventory.add_product(1, "Laptop", 999.99, 10))
print(inventory.add_product(2, "Phone", 599.99, 5))
print(inventory.add_product(3, "Tablet", 299.99, 8))

print(inventory.display_available_products())

print(inventory.update_quantity(1, -2))  # Remove 2 laptops
print(inventory.update_quantity(2, 3))   # Add 3 phones

print(inventory.display_available_products())

```

```python
#6 PROBLEM - 6
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        if self.width is not None and self.height is not None:
            return 2 * (self.length * self.width + self.length * self.height + self.width * self.height)
        else:
            return self.length ** 2  

    def calculate_perimeter(self):
        if self.width is not None and self.height is not None:
            return 4 * (self.length + self.width + self.height)
        else:
            return 4 * self.length  

square = Shape(5)
rectangle = Shape(6, 4, 3)

print(f"Square Area: {square.calculate_area()}")
print(f"Square Perimeter: {square.calculate_perimeter()}")

print(f"Rectangle Area: {rectangle.calculate_area()}")
print(f"Rectangle Perimeter: {rectangle.calculate_perimeter()}")
```

```python
#7 PROBLEM - 7
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        else:
            return 0.0

    def display_student_details(self):
        return f"Student ID: {self.student_id}\nName: {self.name}\nAverage Grade: {self.calculate_average_grade():.2f}"

student1 = Student(1, "vicky", [85, 90, 78, 92, 88])
student2 = Student(2, "vignesh", [76, 88, 91, 82, 79])

print(student1.display_student_details())
print(student2.display_student_details())
```

```python
#8 PROBLEM - 8

class Email:
    def __init__(self, sender, recipient, subject, message):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.message = message

    def send_email(self):
        return f"Email sent from {self.sender} to {self.recipient} with subject: '{self.subject}'"

    def display_email_details(self):
        return f"Sender: {self.sender}\nRecipient: {self.recipient}\nSubject: {self.subject}\nMessage: {self.message}"

email1 = Email("vigneshwaran@example.com", "vicky@example.com", "Meeting Agenda", "Dear vicky, please find the meeting agenda attached.")
email2 = Email("vickyvignesh@example.com", "vignesh@example.com", "Project Update", "Hi vignesh, here is the latest project update.")

print(email1.display_email_details())
print(email1.send_email())

print(email2.display_email_details())
print(email2.send_email())
```

```python
#9 PROBLEM - 9
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        self.posts.append(post_content)
        return "Post added successfully."

    def display_posts(self):
        if self.posts:
            return f"Posts for user '{self.username}':\n" + "\n".join(self.posts)
        else:
            return "No posts available."

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            return f"Matching posts for keyword '{keyword}':\n" + "\n".join(matching_posts)
        else:
            return f"No posts containing keyword '{keyword}' found."

profile = SocialMediaProfile("alice_smith")

print(profile.add_post("Just had a great day at the beach!"))
print(profile.add_post("Working on a new project. #excited"))

print(profile.display_posts())

print(profile.search_posts("beach"))

print(profile.search_posts("food"))
```

```python
#10 PROBLEM 10

class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        self.tasks.append((task, due_date))
        return f"Task '{task}' added with due date {due_date}."

    def mark_task_completed(self, task):
        for i, (t, _) in enumerate(self.tasks):
            if t == task:
                del self.tasks[i]
                return f"Task '{task}' marked as completed."
        return f"Task '{task}' not found."

    def display_pending_tasks(self):
        if self.tasks:
            return "Pending Tasks:\n" + "\n".join([f"Task: {task}, Due Date: {date}" for task, date in self.tasks])
        else:
            return "No pending tasks."

todo = ToDoList()

print(todo.add_task("Buy groceries", "2023-08-15"))
print(todo.add_task("Finish report", "2023-08-20"))

print(todo.display_pending_tasks())

print(todo.mark_task_completed("Buy groceries"))

print(todo.display_pending_tasks())

print(todo.mark_task_completed("Buy groceries"))  
```
