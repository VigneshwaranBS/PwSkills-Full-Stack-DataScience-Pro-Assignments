```python

class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def get_intersection_node(headA, headB):
    if not headA or not headB:
        return None

    # Helper function to get the length of a linked list
    def get_length(node):
        length = 0
        while node:
            length += 1
            node = node.next
        return length

    # Get the lengths of both linked lists
    lengthA = get_length(headA)
    lengthB = get_length(headB)

    # Move the longer linked list's head to make both lists equal in length
    while lengthA > lengthB:
        headA = headA.next
        lengthA -= 1

    while lengthB > lengthA:
        headB = headB.next
        lengthB -= 1

    # Move both pointers until the intersection point is found
    while headA != headB:
        headA = headA.next
        headB = headB.next

    return headA

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create two linked lists with an intersection point: List 1: 1 -> 2 -> 3 -> 4, List 2: 9 -> 8 -> 3 -> 4
common_node = ListNode(3, ListNode(4))
list1 = ListNode(1, ListNode(2, common_node))
list2 = ListNode(9, ListNode(8, common_node))

print("List 1:")
print_linked_list(list1)

print("\nList 2:")
print_linked_list(list2)

# Find the intersection point
intersection_node = get_intersection_node(list1, list2)

if intersection_node:
    print(f"\nIntersection Point: Node with value {intersection_node.value}")
else:
    print("\nNo Intersection Point")
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def reverse_k_groups(head, k):
    # Helper function to reverse a linked list
    def reverse_list(node):
        prev = None
        current = node

        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        return prev

    # Helper function to get the length of the linked list
    def get_length(node):
        length = 0
        while node:
            length += 1
            node = node.next
        return length

    length = get_length(head)
    dummy = ListNode()
    dummy.next = head
    current = dummy

    while length >= k:
        # Find the start and end nodes of the current group
        start = current.next
        end = current.next
        for _ in range(k - 1):
            end = end.next

        # Save the next node after the current group
        next_group = end.next

        # Reverse the current group
        end.next = None
        current.next = reverse_list(start)

        # Connect the reversed group with the next group
        start.next = next_group

        # Move to the next group
        current = start
        length -= k

    return dummy.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

print("Original Linked List:")
print_linked_list(head)

# Reverse nodes in groups of k
k = 3
head = reverse_k_groups(head, k)

print(f"\nLinked List after reversing in groups of {k}:")
print_linked_list(head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def is_palindrome(head):
    # Helper function to reverse a linked list
    def reverse_list(node):
        prev = None
        current = node

        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        return prev

    # Helper function to find the middle of the linked list
    def find_middle(node):
        slow = node
        fast = node

        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

        return slow

    if not head or not head.next:
        return True  # An empty list or a list with a single element is a palindrome

    middle = find_middle(head)
    second_half = reverse_list(middle)

    # Compare the reversed second half with the first half
    first_half = head
    while second_half is not None:
        if first_half.value != second_half.value:
            return False
        first_half = first_half.next
        second_half = second_half.next

    return True

# Example usage:
# Create a linked list: 1 -> 2 -> 2 -> 1
head = ListNode(1, ListNode(2, ListNode(2, ListNode(1))))

print("Linked List:")
while head is not None:
    print(head.value, end=" -> ")
    head = head.next
print("None")

print("\nIs the linked list a palindrome?")
print(is_palindrome(head))
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def rotate_right(head, k):
    if not head or k == 0:
        return head

    # Helper function to get the length of the linked list
    def get_length(node):
        length = 0
        while node:
            length += 1
            node = node.next
        return length

    length = get_length(head)
    k = k % length  # Handle cases where k is greater than the length of the list

    if k == 0:
        return head  # No rotation needed

    # Find the new head and tail positions after rotation
    new_tail = head
    for _ in range(length - k - 1):
        new_tail = new_tail.next

    new_head = new_tail.next
    new_tail.next = None

    # Connect the original tail to the original head to create a circular list
    tail = new_head
    while tail.next:
        tail = tail.next
    tail.next = head

    return new_head

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

print("Original Linked List:")
print_linked_list(head)

# Rotate the linked list to the right by 2 places
k = 2
head = rotate_right(head, k)

print(f"\nLinked List after rotating to the right by {k} places:")
print_linked_list(head)
```

```python
class Node:
    def __init__(self, value=None, prev=None, next=None, child=None):
        self.value = value
        self.prev = prev
        self.next = next
        self.child = child

def flatten_multilevel_list(head):
    if not head:
        return None

    def flatten_helper(node):
        current = node
        tail = node

        while current:
            if current.child:
                child_head, child_tail = flatten_helper(current.child)
                next_node = current.next

                # Connect the current node to the child list
                current.next = child_head
                child_head.prev = current
                current.child = None

                # Connect the child list's tail to the next node
                if next_node:
                    child_tail.next = next_node
                    next_node.prev = child_tail
                    tail = child_tail
                else:
                    tail = child_tail

            else:
                tail = current

            current = current.next

        return node, tail

    flatten_helper(head)

# Function to print the doubly linked list
def print_doubly_linked_list(head):
    current = head
    while current:
        print(current.value, end=" <-> ")
        current = current.next
    print("None")

# Example usage:
# Create a multilevel doubly linked list
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.child = Node(7)
head.next.next.child.next = Node(8)
head.next.next.child.next.prev = head.next.next.child
head.next.next.child.next.child = Node(11)
head.next.next.child.next.child.next = Node(12)
head.next.next.child.next.child.prev = head.next.next.child.next

head.next.next.child.next.child.child = Node(13)

head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(5)
head.next.next.next.next.prev = head.next.next.next
head.next.next.next.next.child = Node(9)
head.next.next.next.next.child.next = Node(10)
head.next.next.next.next.child.next.prev = head.next.next.next.next.child

head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.prev = head.next.next.next.next

print("Original Doubly Linked List:")
print_doubly_linked_list(head)

# Flatten the multilevel doubly linked list
flatten_multilevel_list(head)

print("\nFlattened Doubly Linked List:")
print_doubly_linked_list(head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def rearrange_linked_list(head):
    if not head or not head.next:
        return head

    # Separate odd and even-positioned nodes into two lists
    odd_head = ListNode()
    even_head = ListNode()
    odd_current = odd_head
    even_current = even_head
    current = head
    is_odd = True

    while current:
        if is_odd:
            odd_current.next = current
            odd_current = odd_current.next
        else:
            even_current.next = current
            even_current = even_current.next

        is_odd = not is_odd
        current = current.next

    # Combine the odd and even lists
    odd_current.next = even_head.next
    even_current.next = None

    return odd_head.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

print("Original Linked List:")
print_linked_list(head)

# Rearrange the linked list
head = rearrange_linked_list(head)

print("\nLinked List after rearranging:")
print_linked_list(head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def add_one_to_linked_list(head):
    # Helper function to reverse a linked list
    def reverse_list(node):
        prev = None
        current = node

        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        return prev

    # Reverse the linked list
    reversed_head = reverse_list(head)

    current = reversed_head
    carry = 1  # Adding 1 to the least significant digit

    while current is not None:
        total = current.value + carry
        current.value = total % 10
        carry = total // 10

        if carry == 0:
            break

        current = current.next

    # If there's still a carry, create a new node
    if carry > 0:
        current.next = ListNode(carry)

    # Reverse the linked list back
    result_head = reverse_list(reversed_head)

    return result_head

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage:
# Create a linked list: 1 -> 2 -> 3 (represents the number 123)
head = ListNode(1, ListNode(2, ListNode(3)))

print("Original Linked List:")
print_linked_list(head)

# Add one to the linked list
head = add_one_to_linked_list(head)

print("\nLinked List after adding one:")
print_linked_list(head)
```

```python
def search_insert(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left  # If the target is not found, return the index where it would be inserted

# Example usage:
nums = [1, 3, 5, 6]
target = 5

result = search_insert(nums, target)

print(f"The target {target} should be inserted at index {result}.")
```

```python
def find_min_in_rotated_array(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]

min_element = find_min_in_rotated_array(nums)

print(f"The minimum element in the rotated sorted array is: {min_element}")
```

```python
def search_rotated_array(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid

        # If the left half is sorted
        if nums[left] <= nums[mid]:
            if nums[left] <= target <= nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # If the right half is sorted
        else:
            if nums[mid] <= target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1  # If the target is not found

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]
target = 0

result = search_rotated_array(nums, target)

print(f"The target {target} is found at index {result}.")
```

```python
def count_negative_numbers(grid):
    m, n = len(grid), len(grid[0])
    count = 0
    row, col = 0, n - 1  # Starting from the top-right corner

    while row < m and col >= 0:
        if grid[row][col] < 0:
            # All numbers to the left of the current column are also negative
            count += (col + 1)
            row += 1
        else:
            # Move to the previous column
            col -= 1

    return count

# Example usage:
grid = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]

result = count_negative_numbers(grid)

print(f"The number of negative numbers in the matrix is: {result}")
```

```python
def search_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return False

    rows, cols = len(matrix), len(matrix[0])
    left, right = 0, rows * cols - 1

    while left <= right:
        mid = (left + right) // 2
        mid_value = matrix[mid // cols][mid % cols]

        if mid_value == target:
            return True
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1

    return False

# Example usage:
matrix = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
target = 3

result = search_matrix(matrix, target)

print(f"The target value {target} is present in the matrix: {result}")
```

```python
def find_median_sorted_arrays(nums1, nums2):
    merged_array = sorted(nums1 + nums2)
    total_length = len(merged_array)

    if total_length % 2 == 0:
        # If the total length is even, calculate the average of the middle two elements
        mid_left = merged_array[total_length // 2 - 1]
        mid_right = merged_array[total_length // 2]
        return (mid_left + mid_right) / 2.0
    else:
        # If the total length is odd, return the middle element
        return float(merged_array[total_length // 2])

# Example usage:
nums1 = [1, 3]
nums2 = [2]

result = find_median_sorted_arrays(nums1, nums2)

print(f"The median of the combined sorted array is: {result}")
```

```python
def next_greatest_letter(letters, target):
    left, right = 0, len(letters)

    while left < right:
        mid = (left + right) // 2

        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid

    # After the loop, 'left' will be the index where the target should be inserted
    return letters[left] if left < len(letters) else letters[0]

# Example usage:
letters = ['c', 'f', 'j']
target = 'a'

result = next_greatest_letter(letters, target)

print(f"The smallest letter greater than {target} is: {result}")
```

```python
def sort_colors(nums):
    low, high = 0, len(nums) - 1
    current = 0

    while current <= high:
        if nums[current] == 0:
            nums[low], nums[current] = nums[current], nums[low]
            low += 1
            current += 1
        elif nums[current] == 2:
            nums[current], nums[high] = nums[high], nums[current]
            high -= 1
        else:
            current += 1

# Example usage:
nums = [2, 0, 2, 1, 1, 0]
sort_colors(nums)

print(f"The sorted colors array is: {nums}")
```

```python
def find_kth_largest(nums, k):
    if not nums or k < 1 or k > len(nums):
        return None

    def partition(nums, left, right):
        pivot = nums[right]
        i = left - 1

        for j in range(left, right):
            if nums[j] >= pivot:
                i += 1
                nums[i], nums[j] = nums[j], nums[i]

        nums[i + 1], nums[right] = nums[right], nums[i + 1]
        return i + 1

    def quickselect(nums, left, right, k):
        if left <= right:
            pivot_index = partition(nums, left, right)

            if pivot_index == k - 1:
                return nums[pivot_index]
            elif pivot_index < k - 1:
                return quickselect(nums, pivot_index + 1, right, k)
            else:
                return quickselect(nums, left, pivot_index - 1, k)

    return quickselect(nums, 0, len(nums) - 1, k)

# Example usage:
nums = [3, 2, 1, 5, 6, 4]
k = 2

result = find_kth_largest(nums, k)

print(f"The {k}th largest element is: {result}")
```

```python
def wiggle_sort(nums):
    if not nums:
        return

    n = len(nums)

    for i in range(1, n, 2):
        if i < n - 1 and nums[i] < nums[i + 1]:
            nums[i], nums[i + 1] = nums[i + 1], nums[i]

        if nums[i - 1] > nums[i]:
            nums[i - 1], nums[i] = nums[i], nums[i - 1]

# Example usage:
nums = [3, 5, 2, 1, 6, 4]
wiggle_sort(nums)

print(f"The reordered array is: {nums}")
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage:
input_number = 5
output_factorial = factorial(input_number)

print(f"The factorial of {input_number} is: {output_factorial}")
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
input_number = 7
output_result = is_prime(input_number)

print(f"{input_number} is a prime number: {output_result}")
```

```python
def generate_fibonacci(n):
    fibonacci_series = [0, 1]
    while True:
        next_number = fibonacci_series[-1] + fibonacci_series[-2]
        if next_number <= n:
            fibonacci_series.append(next_number)
        else:
            break
    return fibonacci_series

# Example usage:
input_number = 8
output_series = generate_fibonacci(input_number)

print(f"The Fibonacci series up to {input_number}: {output_series}")
```

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power(base, exponent - 1)

# Example usage:
base_number = 3
exponent_number = 4

result = power(base_number, exponent_number)

print(f"{base_number}^{exponent_number} is: {result}")
```

```python
def reverse_string(s):
    return s[::-1]

# Example usage:
input_string = "hello"
output_reversed = reverse_string(input_string)

print(f"The reversed string of '{input_string}' is: '{output_reversed}'")
```
